```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
Iris = pd.read_csv('iris.csv')
x1=np.array(Iris['sepal_length'])
x2=np.array(Iris['petal_width'])
plt.plot()
plt.title('dataset')
plt.scatter(x1,x2)
plt.show()
x=np.array(list(zip(x1,x2)))
colors=['b','g','r']
markers=['o','v','s']
plt.ylabel('Length')
kmeans=KMeans(n_clusters=3).fit(x)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=200,c='yellow',label='centroids')
for i,l in enumerate(kmeans.labels_):
    plt.plot(x1[i],x2[i],color=colors[l],marker=markers[l])
plt.xlabel('width')
plt.legend()
plt.show()
```

```python
import sklearn.cluster as cluster
import sklearn.metrics as metrics
from sklearn.metrics import davies_bouldin_score
i=3
labels=cluster.KMeans(n_clusters=i).fit(x).labels_
predict=cluster.KMeans(n_clusters=i).fit(x).predict(x)
print("silhouette score for k(clusters)="+str(i)+"is"+str(metrics.silhouette_score(x,labels,metric="euclidean",sample_size=1000)))
print("calinski harabasz score:"+str(metrics.calinski_harabasz_score(x,labels)))
print("davis bouldin score:"+str(davies_bouldin_score(x,predict)))
```
