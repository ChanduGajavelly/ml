```python
import pandas as pd
import numpy as np
import cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
dataset1 = pd.DataFrame(np.random.randn(100, 4), columns=['a', 'b', 'c', 'd'])
dataset1.head()
```

```python
dataset2 = pd.DataFrame({'key': ['a', 'b', 'c'], 'value': [11, 22, 33]})
dataset2.head()
```

```python
dataset3 = pd.DataFrame({'a': [11, 22, 33, 44, 55], 'b': [10, 20, 30, 40, 50], 'c': [5, 4, 3, 2, 1]})
dataset3.head()
```

```python
dataset1.iplot(kind='scatter', x='a', y='c')
```

```python
dataset1.iplot(kind='scatter', x='a', y='c', mode='markers')
```

```python
dataset2.iplot(kind='bar', x='key', y='value')
```

```python
dataset3.iplot(kind='surface')
```

```python
dataset3.iplot(kind='box')
```
