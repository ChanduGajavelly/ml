```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y=np.sin(x)
plt.plot(x,y, marker="x")
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40]
y=[20,25,30,35]
plt.plot(x,y)
plt.title("linear graph" ,fontsize=25, color="red")
plt.show()
```

```python
x=[10,20,30,40]
y=[20,25,30,35]
fig=plt.figure(figsize = (5,4))
ax=fig.add_axes([1,1,1,1])
ax1=ax.plot(x,y)
ax2=ax.plot(y,x)
ax.set_title("linear graph")
ax.set_xlabel("x-axis")
ax.set_ylabel("y-axis")
ax.legend(labels = ('Line 1','Line 2'))
plt.plot(y,x,color='orange',linewidth=3,marker='o',markersize=15,linestyle='--')
plt.plot(x,y,color='green',linewidth=3,marker='o',markersize=15,linestyle='--')
plt.show()
```
